creating empty dataframe for appending all movie details in 122 pages in a single dataframe

```python
df=pd.DataFrame()
```

```python
df
```

```python
warnings.filterwarnings("ignore")
```

```python
for i in range(1,122):
    response = requests.get('https://api.themoviedb.org/3/discover/movie?api_key=bef6355b4dc06bf31b39995bb1b23f19&with_original_language=te&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])
    df = df.append(temp_df,ignore_index=True)[['id','genre_ids','title','overview','poster_path']]
```

```python
df.head()
```

exporting movie details  as pandas dataframe

```python
df.to_csv('TMDB telugu movie details.csv')
```

iterating multiple pages to get details for all movies

```python
cast_details = pd.DataFrame() 
for i in df['id']:
    response = requests.get('https://api.themoviedb.org/3/movie/{}/credits?api_key=bef6355b4dc06bf31b39995bb1b23f19&language=en-US'.format(i))
    temp_df = pd.DataFrame(response.json()['cast'])
    temp_df['id'] = i  
    
    cast_details = cast_details.append(temp_df, ignore_index=True)  


print(cast_details.head())
```
